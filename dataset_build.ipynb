# Import required packages

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#  Reading CSV file 

```python
df = pd.read_csv('./cardekho_dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
pet = 'Petrol' # Petrol
des = 'Diesel' # Diesel
cng = 'CNG' # CNG
dfpet = df.loc[df['fuel_type']== pet] # Filtering data for Petrol
dfdes = df.loc[df['fuel_type']==des] # Filtering data for Diesel
dfcng = df.loc[df['fuel_type']==cng] # Filtering data for CNG
```

#  Plotting the data

Petrol

```python
plt.figure(figsize=(10, 6))
plt.scatter(dfpet['km_driven'], dfpet['selling_price'], color='blue', alpha=0.5)
plt.title('Kilometers Driven vs. Price of Car')
plt.xlabel('Kilometers Driven')
plt.ylabel('Price ($)')
plt.grid(True)
plt.show()
```

Diesel 

```python
plt.figure(figsize=(10, 6))
plt.scatter(dfdes['km_driven'], dfdes['selling_price'], color='blue', alpha=0.5)
plt.title('Kilometers Driven vs. Price of Car')
plt.xlabel('Kilometers Driven')
plt.ylabel('Price ($)')
plt.grid(True)
plt.show()
```

CNG

```python
plt.figure(figsize=(10, 6))
plt.scatter(dfcng['km_driven'], dfcng['selling_price'], color='blue', alpha=0.5)
plt.title('Kilometers Driven vs. Price of Car')
plt.xlabel('Kilometers Driven')
plt.ylabel('Price ($)')
plt.grid(True)
plt.show()
```

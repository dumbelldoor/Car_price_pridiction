# Import required packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
# import tensorflow as tf
from sklearn import preprocessing

```

#  Reading CSV file 

```python
df = pd.read_csv('./cardekho_dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
pet = 'Petrol' # Petrol
des = 'Diesel' # Diesel
cng = 'CNG' # CNG
dfpet = df.loc[df['fuel_type']== pet] # Filtering data for Petrol
dfdes = df.loc[df['fuel_type']==des] # Filtering data for Diesel
dfcng = df.loc[df['fuel_type']==cng] # Filtering data for CNG
```

#  Plotting the data

Petrol

```python
plt.figure(figsize=(10, 6))
plt.scatter(dfpet['km_driven'], dfpet['selling_price'], color='blue', alpha=0.5)
plt.title('Kilometers Driven vs. Price of Car')
plt.xlabel('Kilometers Driven')
plt.ylabel('Price ($)')
plt.grid(True)
plt.show()
```

Diesel 

```python
plt.figure(figsize=(10, 6))
plt.scatter(dfdes['km_driven'], dfdes['selling_price'], color='blue', alpha=0.5)
plt.title('Kilometers Driven vs. Price of Car')
plt.xlabel('Kilometers Driven')
plt.ylabel('Price ($)')
plt.grid(True)
plt.show()
```

CNG

```python
plt.figure(figsize=(10, 6))
plt.scatter(dfcng['km_driven'], dfcng['selling_price'], color='blue', alpha=0.5)
plt.title('Kilometers Driven vs. Price of Car')
plt.xlabel('Kilometers Driven')
plt.ylabel('Price ($)')
plt.grid(True)
plt.show()
```

```python
f_small = df[['km_driven', 'selling_price', 'seller_type', 'fuel_type','transmission_type',]]

corr_matrix = f_small.corr()
```

```python
# Set up the matplotlib figure
plt.figure(figsize=(10, 5))

# Draw the heatmap for the two-column correlation matrix
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=.5)

# Add title
plt.title('Correlation Heatmap between column1 and column2')

# Show the plot
plt.show()
```
